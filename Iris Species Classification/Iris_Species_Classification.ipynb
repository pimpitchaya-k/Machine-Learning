```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import models
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
x = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
x
```

```python
x = x.values
```

```python
mean = x.mean(axis=0)
mean
```

```python
sd = x.std(axis=0)
sd
```

```python
x = (x-mean)/sd
x
```

```python
y = df[['Species']]
y
```

```python
y_dummies = pd.get_dummies(data=y)
y = y_dummies
y
```

```python
y = y.values
y
```

```python
x_train, x_ , y_train, y_ = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
x_test, x_val , y_test, y_val = train_test_split(x_, y_, test_size=0.5, random_state=42, stratify=y_)
```

```python
x.shape
```

```python
inputLayer = layers.Input(shape = (4,) )
```

```python
hypothesis = layers.Dense(3, activation='softmax')(inputLayer)
```

```python
model = tf.keras.Model(inputs = inputLayer, outputs = hypothesis)
```

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model)
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1, momentum=0.95), loss=tf.keras.losses.CategoricalCrossentropy(), metrics=[tf.keras.metrics.CategoricalAccuracy()])
```

```python
earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(x=x_train, y=y_train, batch_size=120, epochs=100, callbacks=[earlystop], validation_data=(x_val, y_val))
```

```python
model.save('model.h5')
```

```python
model.save_weights('weight.h5')
```

```python
y_test_predicted = model.predict(x = x_test)
y_test_predicted
```

```python
y_pred = y_test_predicted.argmax(axis=1)
y_pred
```
